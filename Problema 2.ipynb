```python
import numpy as np
import pandas as pd
#importo paquete para graficas
import pylab as pl
from matplotlib import pyplot as plt
#importo paquetes para calculo simbolico
from sympy.interactive import printing
from sympy import Symbol, Derivative, diff, simplify, solve

#definiciones
av, asup, ac, asym = 16.0, 18.0, 0.72, 28.0
#pido el valor de A por pantalla
print('Valor de A')
A = int(input())
#vectores para la tabla de datos
x = np.zeros(A)
y = np.zeros(A)
```

```python
#defino la energia de ligadura por nucleon
def BA(Z):
    vol = av
    sup = asup*A**(-1/3)
    cul = (ac*A**(-4/3))*(Z**2)
    sym = asym*((A**2)+4*(Z**2)-4*A*Z)/(A**2)
    b = (av-sup-cul-sym)
    return b
```

```python
for i in range(0,A):
    B = 0
    B= BA(i+1)
    x[i] = i+1
    y[i] = B
    
tabla = pd.DataFrame(list(zip(x,y)),columns=["Z","B(Z)/A"])
#tabla
```

```python
plt.figure(figsize=(15,10))
pl.plot (x, y, '.-')
plt.title(f'Energía de ligadura por nucleón para A={A}')
pl.xlabel('Z')
pl.ylabel('B(Z,A)/A[MeV]')
pl.xlim(20,105)
pl.ylim(0,10)
```

```python
# imprimir con notación matemática. 
printing.init_printing(use_latex='mathjax') 
n = Symbol('Z')
#energia de ligadura
Bx = BA(n)
Bx
```

```python
#la derivo
dB = diff(Bx,n)
#calculo el maximo
Zmax = solve(dB)
#tengo que pasarlo a numero, ya que la funcion anterior me devuelve un array
sol = Zmax[0]
sol
```

```python
Bmax=BA(sol)
Bmax
```

```python
pl.figure(figsize=(15,10))
pl.plot (x, y, '.-', label='Energia por nucleon')
pl.plot(sol, Bmax,'o', label=f'Zmax={sol}')
pl.legend()
pl.title(f'Energía de ligadura por nucleón para A={A}')
pl.xlabel('Z')
pl.ylabel('B(Z,A)/A[MeV]')
pl.xlim(20,105)
pl.ylim(0,10)
```
