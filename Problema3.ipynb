```python
import numpy as np
import pylab as pl

import pylab as pl
from matplotlib import pyplot as plt

from sympy.interactive import printing
from sympy import Symbol, Derivative, diff, simplify, solve
av, asup, ac, asym = 16.0, 18.0, 0.72, 28.0
A=250
#Asumo x=A
Qa = np.zeros(A-140)
Qp = np.zeros(A-140)
X = np.zeros(A-140)
```

```python
def Z(x):
    num = 2*asym*x
    den = ac*(x**(2/3))+(4*asym)
    s = (num/den)
    return s
```

```python
def B(x,y):
    vol = av*x
    sup = asup*(x**(2/3))
    cul = ac*(y**2)*(x**(-1/3))
    sim = asym*((x**2)+(4*y**2)-4*x*y)/x
    b = vol-sup-cul-sim
    return b
    
```

```python
def Qalfa(x):
    b = Z(x)
    b1 = B(x-4,b-2)
    b2 = 27.273
    b3 = B(x,b)
    q1 = b1+b2-b3
    return q1
```

```python
def Qprot(x):
    b = Z(x)
    b1 = B(x-1,b-1)
    b3 = B(x,b)
    q2 = b1-b3
    return q2
```

```python
for i in range(140,A):
    q1 = q2 = 0
    q1 = Qalfa(i)
    q2 = Qprot(i)
    
    Qa[i-140] = q1
    Qp[i-140] = q2
    X[i-140] = i
```

```python
plt.figure(figsize=(15,10))
pl.plot (X, Qa, '.-', label='Emision alfa')
pl.plot (X, Qp, '.-', label='Emision proton')
pl.legend()
plt.title(f'Calores de reaccion para distintas emisiones')
pl.xlabel('A')
pl.ylabel('Q[MeV]')
```
